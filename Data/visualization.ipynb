# Data Visualization

Data visualization is the presentation of data in a pictorial or graphical format. This helps us to easily and quickly interpret the data and adjust different variables to see their effect.

## Matplotlib

Matplotlib is a Python package used for 2D graphics. It's a comprehensive library for creating static, animated, and interactive, visualizations in python. It is easily customizable through accessing the class

A simple plot

```python
plt.plot([1,2,3],[4,5,9])
plt.show()
```

A better one

```python
from matplotlib import style 

style.use("ggplot")

x= [10,16,20]
y= [24,32,12]

x2= [12,19,24]
y2= [20,28,18] 

plt.plot(x,y,'g', label ='Line 1', linewidth = 4)
plt.plot(x2,y2,'c', label = "Line 2", linewidth=4)

plt.title('Added Style')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.legend()

plt.grid(True, color = 'k') # Red = "r", Blue = "b", Green = "g", Black = "k"
plt.show()
```

Let's try something else. First generate some data for Gross Domestic Product (GDP) for a period of time

```python
from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
```

Plotting years against gdp

```python
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
plt.title("Nominal GDP")
plt.ylabel("Billions of $")
plt.show()
```

Bar chat with movies data

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 11]
```

```python
plt.bar(range(len(movies)), num_oscars)
plt.title("Favorite moview")
plt.ylabel("# of Academy  Awards")

# label x-axis with movie names at bar centers
plt.xticks(range(len(movies)), movies)
plt.show()
```

Histogram

```python
from collections import Counter
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
```

```python
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)
plt.bar(
    [x + 5 for x in histogram.keys()],
    histogram.values(),
    10,
    edgecolor=(0, 0, 0)
)
plt.axis([-5, 105, 0, 5])
```
