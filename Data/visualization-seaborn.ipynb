# Seaborn

Seaborn is a powerful graphical package that helps us explore and understand data.

It is a library built on top of matplotlib for making statistical graphics in Python. In addition to the bars, pies, lines, scatter plots in matplotlib, it goes further to provide a variety of visualizations and paterns as themes that could easily be integrated into the application

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
import seaborn as sns
```

```python
a = sns.load_dataset("flights")
sns.relplot(x="passengers", y="month", data=a)
```

```python
b = sns.load_dataset("tips")
sns.relplot(x="time", y="tip", data=b, kind ="line")
```

```python
sns.catplot(x="day", y="total_bill", data=b)
```

```python
sns.catplot(x="day", y="total_bill", kind= "violin", data=b)
```

```python
sns.catplot(x="day", y="total_bill",kind="boxen", data=b)
```

```python
a = sns.load_dataset("iris")
b = sns.FacetGrid(a, col = "species")
b.map(plt.hist, "sepal_length")
```
