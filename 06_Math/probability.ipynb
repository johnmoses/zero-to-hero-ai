# Conditional Probability

The events or instances of an experiment may sometimes depend on other events for occurence.  We can that have:
- Independent events
- Dependent events
- Mutually exculisive events

```python
%matplotlib inline
import random
from matplotlib import pyplot as plt

# Create a list with 2 elements of a coin
coin = [0,0]

# Trials
trials = 1000
trial = 0
while trial < trials:
    trial = trial + 1
    # Get a toss for 0 or 1
    toss = random.randint(0,1)
    # Increment the coin
    coin[toss] = coin[toss] + 1

print(coin)

# Plot a pie chart
plt.figure(figsize=(5,5))
plt.pie(coin, labels=['heads', 'tails'])
plt.legend()
plt.show
```

```python
# Math
import math
SQRT_TWO_PI = math.sqrt(2 * math.pi)

def normal_pdf(x: float, mu: float = 0, sigma: float = 1) -> float:
    return (math.exp(-(x-mu) ** 2 / 2 / sigma ** 2) / (SQRT_TWO_PI * sigma))
```

```python
# normal cdf
def normal_cdf(x: float, mu: float=0, sigma: float=1) -> float:
    return (1 + math.erf((x - mu) / math.sqrt(2) / sigma)) / 2
    
```

```python
import matplotlib.pyplot as plt
xs = [x / 10.0 for x in range(-50, 50)]
plt.plot(xs, [normal_pdf(x, sigma=1) for x in xs], '-', label='mu=0, sigma=1')
plt.plot(xs, [normal_pdf(x, sigma=2) for x in xs], '--', label='mu=0, sigma=2')
plt.plot(xs, [normal_pdf(x, sigma=0.5) for x in xs], ':', label='mu=0, sigma=0.5')
plt.plot(xs, [normal_pdf(x, mu=-1) for x in xs], '-,', label='mu=-1, sigma=1')
plt.legend()
plt.title('Various Normal pdfs')
```

## Binomial Variables and Distributions

Binomial variables are random variables that can give the frequency of an event in a fixed number of repeated independent experiments. Let's take for instance, suppose we flip a coin three times, counting heads as success, we can define a binomial variable to represent the number of successful flips.

Here the number of flip events is 3, number of possible outcomes in each flip is 2, then the total number of possible results is 2<sup>3</sup> or 8. Now our variable ***X***, that represents the number of successful events can be 0, 1, 2, or 3

```python
%matplotlib inline
from scipy import special as sps 
from matplotlib import pyplot as plt 
import numpy as np 

trials = 3

possibles = 2**trials 
print(possibles)
x = np.array(range(0, trials+1))
p = np.array([sps.comb(trials, i, exact=True)/possibles for i in x])

# Plot the graph
plt.xlabel('Successes')
plt.ylabel('Probabilities')
plt.bar(x,p)
plt.show()
```
