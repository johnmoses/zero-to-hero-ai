# K-Means Clustering

```python
import sys; sys.path.append("..")
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

Datasets

```python
from sklearn.datasets import make_blobs

np.random.seed(42)

X, y = make_blobs(
    centers=3, n_samples=500, n_features=2, shuffle=True, random_state=40
)
print(X.shape)

clusters = len(np.unique(y))
print(clusters)
```

Fit data

```python
from xalgos.k_means_1 import KMeans

k = KMeans(K=clusters, max_iters=150, plot_steps=False)
k_pred = k.predict(X)

print("Accuracy: ",
      accuracy_score(y, k_pred))
print("Precision: ",
      precision_score(y, k_pred, average='weighted'))
print("Recall: ",
      recall_score(y, k_pred, average='weighted'))
print("F1-Score: ",
      f1_score(y, k_pred, average='weighted'))
```
