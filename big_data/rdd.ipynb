# Resilient Distributed Datasets

Resilient Distributed Datasets (RDD) is the basic abstraction in Spark. It is an immutable distributed collection of objects. RDD is a fundamental and backbone data type in Spark.

```python
import pyspark as spark
```

```python
sc = spark.SparkContext.getOrCreate()
```

## RDD from list of words

```python
RDD = sc.parallelize(["Spark", "is", "a", "framework", "for", "Big Data processing"])
print("The type of RDD is", type(RDD))
```

## RDD from external datasets

```python
file_path = '../xdata/5000_points.txt'
```

```python
fileRDD = sc.textFile(file_path)
print("The file type of fileRDD is", type(fileRDD))
```

Partitions in the data

```python
print("Number of partitions in fileRDD is", fileRDD.getNumPartitions())

# Create a fileRDD_part from file_path with 5 partitions
fileRDD_part = sc.textFile(file_path, minPartitions = 5)

# Check the number of partitions in fileRDD_part
print("Number of partitions in fileRDD_part is", fileRDD_part.getNumPartitions())
```

map() Transformation

```python
RDD = sc.parallelize([1,2,3,4])
RDD_map = RDD.map(lambda x: x * x)
RDD_map
```

```python
RDD = sc.parallelize(["hello world", "how are you"])
RDD_flatmap = RDD.flatMap(lambda x:x.split(" "))
```
