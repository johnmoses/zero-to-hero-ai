# Data Pre-processing

```python
%matplotlib inline
import pandas as pd
import numpy as np
```

## Standard Scaller

Standardization is a preprocessing method used to transform continuous data to make it look normally distributed.
This is needed in the following scenarios
* Models in linear space
* Dataset features have high variance 
* If data is not normally distributed, this is not the best scaler to use.

The StandardScaler assumes your data is normally distributed (continuous numerical data) within each feature and will scale them such that the distribution is now centred around 0, with a standard deviation of 1.

Generate normally distributed data

```python
df = pd.DataFrame({
    'x1': np.random.normal(0, 2, 10000),
    'x2': np.random.normal(5, 3, 10000),
    'x3': np.random.normal(-5, 5, 10000)
})
```

```python
df.plot.kde()
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
data_tf = ss.fit_transform(df)
```

```python
df = pd.DataFrame(data_tf, columns=['x1','x2','x3'])
```

```python
df.plot.kde()
```

## MiniMaxScaler

This is another scaller necessary in instances where:

* Need to calculate - Subtract min of column & div by difference between max & min
* Data shifts between 0 & 1
* Distribution is not suitable for StandardScaler.
* Need sensitivity to outliers

```python
df = pd.DataFrame({
    # positive skew
    'x1': np.random.chisquare(8, 1000),
    # negative skew 
    'x2': np.random.beta(8, 2, 1000) * 40,
    # no skew
    'x3': np.random.normal(50, 3, 1000)
})
```

```python
df.plot.kde()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
data_tf = mm.fit_transform(df)
```

```python
df = pd.DataFrame(data_tf,columns=['x1','x2','x3'])
df.plot.kde()
```
